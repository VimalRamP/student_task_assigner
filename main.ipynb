```python
from my_pack import sta #import the package of user defined functions
import csv

class task_assigner: #create a class for login function

    def load_users_from_csv(self, filename): #function defined for reading file 
        users = {}
        with open(filename, "r") as file:
            reader = csv.reader(file)
            for row in reader:
                username, password = row
                users[username] = password
        return users
        
    def teacher_login(self): #function defined for checking teacher user name & password
        users = self.load_users_from_csv("teacher.csv")

        if users is None:
            print("Error loading user data. Please check the CSV file.")
            return False

        username = input("Enter teacher username: ")
        password = input("Enter teacher password: ")

        if username in users and password == users[username]:
            print(f"Teacher {username} logged in successfully.")
            return True
        else:
            print("Invalid username or password. Please try again.")
            x=task_assigner()
            x.teacher_login()

    def student_login(self): #function defined for checking student user name & password
        users = self.load_users_from_csv("student.csv")

        if users is None:
            print("Error loading user data. Please check the CSV file.")
            return False

        username = input("Enter student username: ")
        password = input("Enter student password: ")

        if username in users and password == users[username]:
            print(f"student {username} logged in successfully.")
            return True
        else:
            print("Invalid username or password. Please try again.")
            y=task_assigner()
            y.student_login()
 

def main(): #defining a main funtion to perform operations
    tasks = []
    STA = task_assigner() #create an object for class

    while True: #looping for login as teacher or student
        print("\nTask Manager Menu:")
        print("1. Teacher Login")
        print("2. Student Login")
        print("0. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            STA.teacher_login()
            while True: #looping to perform teacher operations
                print("\n*** Student Task Assigner ***")
                print("1. Add Task")
                print("2. View Tasks")
                print("3. Remove Task")
                print("4. Edit Task")
                print("5. Clear All Tasks")
                print("6. Search & Filter Task")
                print("7. Export csv")
                print("0. Exit")

                choice = input("Enter your choice: ")
                if choice == '1':
                    sta.add_task(tasks) #calling a function to add task
                elif choice == '2':
                    sta.view_tasks(tasks) #calling a function to view task
                elif choice == '3':
                    sta.remove_task(tasks) #calling a function to remove task
                elif choice == '4':
                    sta.edit_task(tasks) #calling a function to edit task
                elif choice == '5':
                    sta.clear_all_tasks(tasks) #calling a function to clear all task
                elif choice == '6':
                    sta.search_tasks(tasks) #calling a function to search or filter task
                elif choice == '7':
                    sta.export_csv(tasks) #calling a function to export task
                elif choice == '0':
                    print("Exiting the * Student Task Assigner *.")
                    break
                else:
                    print("Invalid choice. Please try again.")

        elif choice == '2':
            STA.student_login()
            while True: #looping to perform student operations
                print("\n*** Student Task Assigner ***")
                print("1. View Tasks")
                print("2. Set Reminder")
                print("3. Mark Task as Done")
                print("4. Search & Filter Task")
                print("5. Export csv")
                print("0. Exit")

                choice = input("Enter your choice: ")
                if choice == '1':
                    sta.view_tasks(tasks) #calling a function to view task
                elif choice == '2':
                    sta.set_reminder(tasks) #calling a function to set remainder to a task
                elif choice == '3':
                    sta.mark_task_done(tasks) #calling a function to mark task as done
                elif choice == '4':
                    sta.search_tasks(tasks) #calling a function to search or filter task
                elif choice == '5':
                    sta.export_csv(tasks) #calling a function to export task
                elif choice == '0':
                    print("Exiting the * Student Task Assigner *.")
                    break
                else:
                    print("Invalid choice. Please try again.")

        elif choice == '0':
            print("Exiting the Student Task Assigner...")
            break
        else:
            print("Invalid choice. Please try again.")
       

if __name__ == "__main__":
    main()
```
